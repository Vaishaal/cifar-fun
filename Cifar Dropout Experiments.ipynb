```python
def make_gaussian_dropout_filter_gen(bandwidth, dropout=0.5, patch_size=6, channels=3):
    ps = patch_size
    def gaussian_filter_gen(num_filters):
        out = np.random.randn(num_filters, channels, ps, ps).astype('float32') * bandwidth
        bitmask = numpy.random.binomial(1, dropout, out.shape)
        dropped_out = np.sum(bitmask)
        total = len(bitmask.ravel())
        scale = dropped_out/float(total)
        print "SCALE: ", scale
        return bitmask * (out/scale)
    return gaussian_filter_gen

def make_gaussian_filter_gen(bandwidth, patch_size=6, channels=3):
    ps = patch_size
    def gaussian_filter_gen(num_filters):
        out = np.random.randn(num_filters, channels, ps, ps).astype('float32') * bandwidth
        return out
    return gaussian_filter_gen
```
